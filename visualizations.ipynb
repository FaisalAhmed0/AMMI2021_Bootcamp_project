```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "train_data.txt"
```

```python
data_df = pd.read_csv(path, sep=',', names = ['ID', 'f1', 'f2', 'f3', 'f4', 'f5', 'f6', 'f7', 'f8', 'f9', 'f10', 'f11', 'class'])
```

```python
data_df.head()
```

```python
# Task 2
fig = plt.figure(figsize=(12, 6))
class_data = data_df['class']
plt.hist(class_data)
plt.xlabel('Class', fontdict={'size': 20})
plt.title("Class Column Distribution",  fontdict={'size': 20})
```

```python
column = data_df['f1']
plt.scatter(column, data_df['f2'], c=data_df['class'], cmap='viridis')
```

```python
data_df.columns
```

```python
def relationship_plot(df, column_name):
    columns = df.columns
    x_col = df[column_name]
    fig = plt.figure(figsize=(18,15))
    for i in range(1, 12):
        plt.subplot(4, 3, i)
        plt.scatter(x_col, df[columns[i]], c= data_df['class'])
```

```python
relationship_plot(data_df, 'f1')
```
